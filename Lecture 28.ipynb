## Random Processes

Randomness in systems make them unpredictable. In order to model random processes, we need to be able to incorporate an element of randomness in our computer programs. Some truly unpredictable or random processes in physics include radioactive decay, quantum tunelling, and basically anything related to quantum mechanics. Some apparently random processes include Brownian motion, genetic mutation, weather patterns, etc., where randomness is used as a conveninent modelling tool. Whether true randomness exists in nature, or whether what we experience as random is either a lack of understanding or unavoidable errors due to limitations of our measurement devices, is a matter of debate. Regardless, we embark upon the journey of understanding and modelling random processes.

## Random Numbers

Random numbers in computers are at best pseudorandom numbers. They are generated by deterministic formula in "random number generators".  

Consider the equation:
$$ x^\prime = (ax+c) \% m$$
where $a$, $c$ and $m$ are integer constants, and $x$ is an integer variable. Using a "seed" $x$, if we keep generating $x^\prime$ and replace $x$ with the new $x^\prime$, let us see what happens then.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
a = 1664525
c = 1013904223
m = 4294967296
x = 1
results = []

for i in range(N):
    x = (a*x+c)%m
    results.append(x)
plt.plot(results,"o")
plt.show()
```

This is a **linear congruential random number generator** (or LCG). It is completely deterministic -- no matter how many time you run this code, it will produce exactly the same plot. The values always lie in the range of 0 to $m-1$.  
### Can you modify the above code to generate "random" numbers between 0 and 1?

```python
N = 100
a = 1664525
c = 1013904223
m = 4294967296
x = 1
results = []

for i in range(N):
    x = (a*x+c)%m
    results.append(x)
plt.plot(results,"o")
plt.show()
```

### Why these values for $a$, $c$ and $m$?

These are tried and tested values, that provide maximum randomness given that we start with a deterministic equation. Any combination of $a$, $c$ and $m$ **would not work**.  
$$ \begin{align}
{\rm the\ modulo}&\qquad m>0;\\
{\rm the\ multiplier}&\qquad 0<a<m;\\
{\rm the\ increment}&\qquad 0 \leq c<m;\\
{\rm the\ seed}&\qquad 0\leq x<m 
\end{align}$$

### Major Drawback

A serious drawback of the linear congruential random number generator is that there are correlations between values of successive numbers. True random numbers should be uncorrelated. 

### Mersenne Twister

It uses a Mersenne prime as its period length. It is a "generalized feedback shift-register generator". It is very complicated to code, but python has built-in functions in the `random` package:  
* `random()`: Gives a random floating point number uniformly distributed in the range from 0 to 1, including 0 but not including 1. 
* `randrange()`: Gives a random integer from 0 to $n-1$.
* `randrange(m,n)`: Gives a random integer from $m$ to $n-1$.
* `randrange(m,n,k)`: Gives a random integer from $m$ to $n-1$, in steps of $k$.

```python
import random

for i in range(10):
    print(random.random())
```

```python
for i in range(10):
    print(random.randrange(10))
```

### Random Number Seed

While for LCG the seed is an input value that tells the generator where to start its sequence, for a Mersenne twister the seed specifies the starting point indirectly and is actually the first value in the sequence. Python has a function `seed` that takes an **integer seed** as an argument.

```python
# seed tells the program where to start the random number sequence
# without a seed, you get a different set of random numbers
# seed is useful for debugging code with random numbers
random.seed(11)
for i in range(4):
    print(random.randrange(10))
```
